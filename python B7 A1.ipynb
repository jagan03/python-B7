```
#LIST FUNCTIONS
list=["python","Java","C","HTML"]
print(list)
print(list[1])
list.append("C++")
print(list)
list.remove("C")
print(list)
list.index("C++")
print(list)
list.reverse()
print(list)
list.append("game")
print(list)
list.sort(key=len)
print(list)

#DICTIONARY FUNCTIONS
my_dict={}
my_dict={1:"python",2:"java"}
print(my_dict)
my_dict={"program":"python",1:[2,3,4]}
print(my_dict)
print(my_dict["program"])
my_dict.pop(1)
print(my_dict)
my_dict.values()

#SETS FUNCTIONS
my_set={1,2,3}
print(my_set)
my_set={7.7,"python",(1,2,3)}
print(my_set)
my_set={1,2,3,2,3,4} #ignores duplicate values
print(my_set)
my_set.add(5)
print(my_set)
my_set.update([6,7])
print(my_set)
my_set.discard(7)
print(my_set)
my_set.remove(6)
print(my_set)
my_set.pop()
my_set.clear()

#TUPLES METHODS
my_tuple=()
print(my_tuple)
my_tuple=(1,2,3)
print(my_tuple)
my_tuple=("python", [4,5,6], (1, 2, 3))
print(my_tuple)
my_tuple=('p','y','t','h','o','n')
print(my_tuple[0])
print(my_tuple[1:5])
print(my_tuple[2:5])
print(my_tuple[-4])
print(my_tuple.count('h'))
print(my_tuple.index('o'))

#STRING METHODS
string="i love programming"
string.capitalize()
print(string)
print(string.isupper())
string="jagadeesh"
print(string.isalpha())
string="77865"
print(string.isdigit())
string="1/2"
print(string.isdecimal())
```
